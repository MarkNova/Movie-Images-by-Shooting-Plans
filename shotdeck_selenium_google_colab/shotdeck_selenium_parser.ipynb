```python
frame_types = [
              #  'Extreme%20Wide',
              #  'Wide',
              #  'Medium%20Wide',
              #  'Medium',
               'Medium%20Close%20Up',
               'Close%20Up',
               'Extreme%20Close%20Up']
```

```python
import os

dir_path = '/content/drive/MyDrive/frames/low_w_frames/'  # Замените пути к папкам на ваши собственные

file_names = []

for frame_path in frame_types:
    folder_path = dir_path + frame_path

    for file_name in os.listdir(folder_path):
        file_names.append(file_name[:-4])
```

```python
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
driver = web_driver()
driver.get('https://shotdeck.com/welcome/home')
driver.maximize_window()


mail_reg = driver.find_elements(By.CSS_SELECTOR, 'input.form-control')[0]
pas_reg = driver.find_elements(By.CSS_SELECTOR, 'input.form-control')[1]
confirm_btn = driver.find_element(By.CSS_SELECTOR, 'button.btn.btn-primary')

mail_reg.send_keys(email)
pas_reg.send_keys(password)
confirm_btn.click()

driver.get('https://shotdeck.com/browse/stills')

driver.find_element(By.CSS_SELECTOR, 'div.accordion-item.frame_size').click()

frame_sizes = driver.find_element(By.CSS_SELECTOR, 'div.accordion-item.frame_size').find_element(By.CSS_SELECTOR, 'div.accordion-body').find_elements(By.CLASS_NAME, 'filter-group')

frame_sizes[0].click()
```

```python
from tqdm import tqdm
```

```python
url = 'https://shotdeck.com/browse/stills#/frame_size/'
for frame_type in tqdm(frame_types):
    t_url = url + frame_type

    driver.get(t_url)

    c = 0
    speed = 3000
    id_list = []
    while len(id_list) < 5500:

        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div.outerimage.jg-entry.entry-visible')))

        image_windows = driver.find_elements(By.CSS_SELECTOR, 'div.outerimage.jg-entry.entry-visible')

        for i in range(len(image_windows)):

            j = image_windows[i]

            img_id = j.get_attribute('id')

            if (img_id not in id_list) and (img_id not in file_names):

                id_list += [img_id]
                print(img_id[5:])
                print(len(id_list))




        driver.execute_script(f'window.scrollTo({c * speed}, {c * speed + speed})')
        c += 1
        speed = speed * 1.5
    for i in id_list:

        img = f'https://shotdeck.com/assets/images/stills/{i[5:]}.jpg'
        driver.get(img)
        image_path = f'/content/drive/MyDrive/frames/low_w_frames/{frame_type}/{i[5:]}.jpg'
        driver.save_screenshot(image_path)
        image = Image.open(image_path)
        os.remove(image_path)
        resized_image = image.convert('RGB')
        resized_image = resized_image.resize((768, 432))
        resized_image.save(image_path)
```
